```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
df=pd.read_excel('online_retail.xlsx')  
df.head(5)
```

```python
df['Description']=df['Description'].str.strip()
df.dropna(axis=0,subset=['Invoice'],inplace=True)
df['Invoice']=df['Invoice'].astype('str')
df=df[~df['Invoice'].str.contains('C')]
df.head(5)
df.shape
```

```python
#after cleaning we need to consolidate the item into one transaction per row
#with each product for the sake of keeping the dataset samll
#we are considering sales for france here
basket=(df[df["Country"]=='France'].groupby(
               ['Invoice','Description'])['Quantity']
               .sum().unstack().reset_index().fillna(0)
       .set_index('Invoice'))
basket
```

```python
basket[50:100]
```

```python
# there are many zeros in the data.we have to make sure that any positive 
#number converted to 1,
#anything less than zero is converted to zero
def encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
basket_sets=basket.applymap(encode_units)
basket_sets.drop("POSTAGE",inplace=True,axis=1)
#basketlist = list(basket_sets)
#basketlist.extend(100*[basketlist[0]])
basket_sets
```

```python
##genrating rules corresponding support,confidence,and lift
frequent_items=apriori(basket_sets,min_support=0.1,use_colnames=True)
rules=association_rules(frequent_items,metric="lift",min_threshold=1)
rules.head()
```
